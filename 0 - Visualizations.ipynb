```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = pd.read_csv('train.csv')
```

```python
dt.head()
```

```python
def rotate_labels(labels, deg):
    for label in labels:
        label.set_rotation(45)
```

```python
f, (ax2) = plt.subplots(1, 1, figsize=(16, 4))
sns.countplot(x="AnimalType", data=dt,  ax=ax2)
```

```python
f, (ax1) = plt.subplots(1, 1, figsize=(20, 4))
gr = sns.countplot(x="AgeuponOutcome", data=dt,  ax=ax1)
rotate_labels(gr.get_xticklabels(), 45)
```

```python
f, (ax1) = plt.subplots(1, 1, figsize=(16, 4))
sns.countplot(x="OutcomeType", hue="AnimalType", data=dt,  ax=ax1)
```
